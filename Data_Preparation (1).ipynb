## Loading normal_man Dataset

```python
from google.colab import files

# Demander à l'utilisateur de sélectionner les fichiers audio à importer
uploaded = files.upload()

# Les fichiers seront stockés dans un dictionnaire où les clés sont les noms des fichiers et les valeurs sont les contenus des fichiers
# Get the filename of the uploaded file
file_name = next(iter(uploaded))

# Access the file contents
file_contents = uploaded[file_name]

file_contents
```

```python
import pandas as pd
import librosa
import io
# Create an empty DataFrame
df1 = pd.DataFrame(columns=['File Name', 'Duration', 'Sampling Rate'])

# Iterate over uploaded files
for file_name, file_contents in uploaded.items():
    # Load audio data
    audio_data, sampling_rate = librosa.load(io.BytesIO(file_contents), sr=None)

    # Calculate duration of the audio in seconds
    duration_seconds = len(audio_data) / sampling_rate


    # Append data to DataFrame
    df1 = df1.append({'File Name': file_name,
                    'Duration': duration_seconds,
                    'Sampling Rate': sampling_rate,
                      'Audio Data': audio_data}, ignore_index=True)
```

```python
# Display the DataFrame
print(df1)
```

```python
# Set 'pitch' to 'high' for filenames containing 'high' or 'h'
df1.loc[df1['File Name'].str.contains('normal|n', case=False), 'pitch'] = 'normal'

# Set 'pitch' to 'high' for filenames containing 'high' or 'h'
df1.loc[df1['File Name'].str.contains('high|h', case=False), 'pitch'] = 'high'

# Set 'pitch' to 'low' for filenames containing 'low' or 'l'
df1.loc[df1['File Name'].str.contains('low|l', case=False), 'pitch'] = 'low'
```

```python
# Add a column 'case' with value 'normal'
df1['case'] = 'normal'
df1['gender'] = 'man'
# Print the DataFrame to verify the changes
print(df1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Univariate analysis
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(df['Duration'], kde=True)
plt.title('Duration Distribution')

plt.subplot(2, 2, 2)
sns.histplot(df['pitch'], kde=True)
plt.title('Pitch Distribution')

plt.subplot(2, 2, 3)
sns.countplot(data=df, x='gender')
plt.title('Gender Distribution')

plt.subplot(2, 2, 4)
sns.countplot(data=df, x='case')
plt.title('Dysphonia Distribution')

plt.tight_layout()
plt.show()

# Bivariate analysis
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Duration', y='pitch', hue='gender')
plt.title('Duration vs Pitch by Gender')

plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='case', y='pitch')
plt.title('Pitch Distribution by Dysphonia')
plt.show()

# Correlation analysis
corr_matrix = df[['Duration', 'pitch', 'Sampling Rate']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
